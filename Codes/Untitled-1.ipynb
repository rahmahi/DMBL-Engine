```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-np.pi, np.pi, 100)
y = np.sin(x)
plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("Angle [radians]")
plt.ylabel("sin(x)")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi, 100)
y = np.sin(x)

plt.plot(x, y)

# Set ticks at multiples of pi/2
ticks = [-np.pi, -np.pi/2, 0, np.pi/2, np.pi]
labels = [r"$-\pi$", r"$-\frac{\pi}{2}$", r"$0$", r"$\frac{\pi}{2}$", r"$\pi$"]

plt.xticks(ticks, labels)

plt.title("Sine Wave")
plt.xlabel("Angle [radians]")
plt.ylabel("sin(x)")
plt.grid()
plt.show()
```
